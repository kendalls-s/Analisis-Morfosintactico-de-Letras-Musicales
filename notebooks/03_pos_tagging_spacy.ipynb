# Importación de Librerías y Carga del Modelo spaCy


```python
import pandas as pd
import spacy

# Cargar modelo de spaCy
nlp = spacy.load("en_core_web_sm")

print("Modelo cargado correctamente")
```

# Carga del Dataset


```python
df = pd.read_csv(r"/songs_with_emotions.csv")

df.head()
```

# Tokenización

```python
df["Tokens"] = df["Clean_Lyrics"].apply(
    lambda text: [token.text for token in nlp(str(text))]
)
```

# POS Taggin


```python
df["POS"] = df["Clean_Lyrics"].apply(
    lambda text: [(token.text, token.pos_) for token in nlp(str(text))]
)

```

# Stop words

```python
df["No_StopWords"] = df["Clean_Lyrics"].apply(
    lambda text: [
        token.text for token in nlp(str(text))
        if not token.is_stop and not token.is_punct
    ]
)
```

# NER

```python
df["Entities"] = df["Clean_Lyrics"].apply(
    lambda text: [(ent.text, ent.label_) for ent in nlp(str(text)).ents]
)
```

# Normalizacion

```python
df["Lower_Text"] = df["Clean_Lyrics"].str.lower()
```

# Lematización

```python
df["Lemmas"] = df["Clean_Lyrics"].apply(
    lambda text: [
        token.lemma_ for token in nlp(str(text))
        if not token.is_punct
    ]
)
```
