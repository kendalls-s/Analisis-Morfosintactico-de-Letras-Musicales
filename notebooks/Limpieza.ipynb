```python
# =========================
# IMPORTAR LIBRERIAS
# =========================

import pandas as pd
import re
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Descargar diccionario VADER (solo primera vez)
nltk.download('vader_lexicon')


# =========================
# 1. CARGAR DATASET
# =========================

ruta_archivo = "C:\\Users\\Roberto\\Downloads\\test.csv\\test.csv"
df = pd.read_csv(ruta_archivo)



# =========================
# 3. ANALISIS DE EMOCIONES
# =========================

sia = SentimentIntensityAnalyzer()

def detectar_emocion(texto):

    scores = sia.polarity_scores(str(texto))

    compound = scores['compound']
    pos = scores['pos']
    neg = scores['neg']

    if compound >= 0.6:
        return "Happy"
    elif 0.3 <= compound < 0.6:
        return "Excited"
    elif compound <= -0.6:
        return "Angry"
    elif -0.6 < compound <= -0.3:
        return "Sad"
    else:
        return "Neutral"


# Crear columna Emotion usando texto limpio
df["Emotion"] = df["Clean_Lyrics"].apply(detectar_emocion)


# =========================
# 4. GUARDAR DATASET FINAL
# =========================


df.to_csv("songs_with_emotions.csv", index=False)


# =========================
# 5. MOSTRAR RESULTADOS
# =========================

print(df.head())
print("\nCantidad de canciones por emociÃ³n:")
print(df["Emotion"].value_counts())

```
